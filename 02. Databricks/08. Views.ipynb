# Prerequisites

## Load employee.csv into DataFrame

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
        header=True,
        inferSchema=True,
        sep = '|',
        quote="'",
    ).limit(10)
df.display()
```

# Views

## Temp Views

### Create Temp View

```python
df.createTempView("user_tmp_vw")
```

### Access Temp View

```python
%sql

SELECT * FROM user_tmp_vw
```

### List Temp Views

```python
for view in spark.catalog.listTables():
    print(f"{view.name} - {view.namespace}")
```

```python
%sql

SHOW VIEWS;
```

## Global Views

### Create Global View

```python
df.createGlobalTempView("user_glb_vw")
```

### Access Global View

```python
%sql

SELECT * FROM global_temp.user_glb_vw
```

### List Global Views

```python
for view in spark.catalog.listTables("global_temp"):
  print(f"{view.name} - {view.namespace}")
```

```python
%sql

SHOW VIEWS in global_temp;
```

```python
for view in spark.catalog.listTables("global_temp"):
  if len(view.namespace) == 0:
      print("Temporary View: ", view.name)
  else:
      print("Global View: ", view.name)
```

```python
for view in spark.catalog.listTables("global_temp"):
  view_type = "Temporary View: " if (len(view.namespace) == 0) else "Global View: "
  print(f{view_type}, view.name)
```
