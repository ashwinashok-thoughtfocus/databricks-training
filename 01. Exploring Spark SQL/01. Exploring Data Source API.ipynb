# Custom Schema

```python
from pyspark.sql.types import *

USER_SCHEMA = StructType([
    StructField("id", IntegerType()),
    StructField("age", IntegerType()),
    StructField("gender", StringType()),
    StructField("designation", StringType()),
    StructField("salary", IntegerType()),
])

df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    schema=USER_SCHEMA
)
df.printSchema()
df.display()
```

# DataFrame Writer API

## Convert CSV into JSON

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True
)
type(df.write)
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
)
```

```python
spark.read.json(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json").show()
```

## Output Modes

### ErrorIfExists

```python
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
)
```

### Overwrite

```python
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
    mode="overwrite"
)
```

### Append

```python
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
    mode="append"
)
```

### Ignore

```python
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
    mode="ignore"
)
```
