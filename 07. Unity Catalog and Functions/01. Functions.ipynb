```python
%sql
 
USE CATALOG quickstart_catalog;
USE SCHEMA quickstart_schema;
CREATE TABLE IF NOT EXISTS users_int(
  id INT,
  first_name STRING,
  last_name STRING
);
DESCRIBE FORMATTED users_int;
```

```python
%sql
DESCRIBE FORMATTED users_int;
```

```python
%sql
INSERT INTO  users_int VALUES(1,"A","B");
INSERT INTO  users_int VALUES(1,"C","D");
```

```python
%sql
CREATE FUNCTION quickstart_catalog.quickstart_schema.full_name_udf(
    first_name STRING, last_name STRING
  )
  RETURNS STRING
  RETURN CONCAT(first_name, "-", last_name)
```

```python
%sql
SELECT full_name_udf(first_name,last_name) FROM  users_int
```
