```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
```

# Writing into parquet format

```python
df.write.format("parquet").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
)
spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
).display()
```

```python
df_parquet = spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
)
df_parquet.display()
```

## Create view

```python
df_parquet.createOrReplaceTempView("users_vw")
```

## Disadvantage - Updates are not allowed

```python
%sql

UPDATE users_vw
SET country = 'Bharat'
WHERE country = 'India'
```

## Disadvantage - Overwrite existing directory no support for roll back

```python
from pyspark.sql.functions import col

df.filter(col("country") == "India").write.mode("overwrite").format("parquet").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
)
```
