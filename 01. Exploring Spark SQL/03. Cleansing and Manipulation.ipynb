# Prerequisites

## Load employees.csv into DataFame

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
        sep="|",
        inferSchema=True,
        header=True,
        quote="'",
    )
display(df)
```

# Dropping Missing Records

```python
df.na.drop().display()
```

```python
df.na.drop(subset=["id","name"]).display()
```

```python
df.na.drop(how="all").display()
```

```python
from pyspark.sql.functions import col
df.filter((col("id").isNotNull()) & (col("name").isNotNull())).display()
```

# Filling Missing Records

```python
df.na.fill(0).display()
```

```python
df.na.fill(0).na.fill("Not in Source").display()
```

```python
df.na.fill({
    "id": 0,
    "name": "Not in Source",
    "exp": 0,
    "gen": "NA",
}).display()
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id":-1,
    "exp":int(mean),
    "gen":"NA"
}).display()
```
