# Bronze Layer
medallion_staging -> Bronze Table

## Customers

```python
spark.read.csv(
    path="/Volumes/dev_catalog/default/retail_staging/customers/customers.csv",
    header=True,
    inferSchema=True,
).write.format("delta").mode("overwrite").saveAsTable(
    "dev_catalog.bronze.customers"
)
```

```python
spark.read.table("dev_catalog.bronze.customers").display()
```
