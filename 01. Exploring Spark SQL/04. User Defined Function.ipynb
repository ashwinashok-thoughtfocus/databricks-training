# Prerequisites

## Load employee.csv into DataFrame

```python
employee_df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
    sep="|",
    quote="'"
).limit(8)
employee_df.display()
```

# Applying python function inside Pyspark

```python
def convert_to_title(st):
    if st is None:
        return None
    return st.title()
```

```python
employee_df.select("id", convert_to_title("name")).display()
```

# Use Python Function

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
convert_to_title_udf = udf(lambda e: convert_to_title(e) if e is not None else None, StringType())
```

```python
employee_df.select("id", convert_to_title_udf("name")).display()
```

```python
employee_df.createOrReplaceTempView("employee_vw")
```

```python
%sql

SELECT id, convert_to_title_udf(name) from employee_vw LIMIT 10
```

```python
spark.udf.register("convert_to_title_udf", convert_to_title, StringType())
```

### Using Spark SQL

```python
spark.sql("SELECT id, convert_to_title(name) FROM employee_vw").display()
```

### Using SQL

```python
%sql

SELECT id, convert_to_title_udf(name) from employee_vw LIMIT 10
```
